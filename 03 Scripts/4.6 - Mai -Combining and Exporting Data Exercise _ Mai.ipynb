```python
# Import libraries

import pandas as pd
import numpy as np
import os
```

```python
# Import Data
path = r'/Users/maitran/Documents/Instacart Basket Analysis'
```

```python
# Import the orders_products_combined dataframe
df_ords_prods_combined = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_combined.pkl'))
```

```python
# Import the df_prods_checked dataframe
df_prods_checked = pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'products_checked.csv'), index_col = False)
```
