### 1. If-Statements with User-Defined Functions

```python
# Define the price_label function
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Apply the function
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
# Check for most expensive product within the subset
df['prices'].max()
```

### 2. If-Statements with the loc() Function

```python
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df['price_range_loc'].value_counts(dropna = False)
```

#### Question 2: Create new colum for “Busiest days”

```python
# Which day most orders take place
ords_prods_merged['order_day_of_week'].value_counts(dropna = False)
```

#### Observations: The two busiest days are 0 (Saturday) and 1 (Sunday). The two slowest days are 4 (Wednesday) and 3 (Tuesday)

```python
result = []

for value in ords_prods_merged["order_day_of_week"]:
  if value == 0 or value == 1:
    result.append("Busiest days")
  elif value == 3 or value == 4:
    result.append("Slowest days")
  else:
    result.append("Regularly days")
```

```python
# Create a new column
ords_prods_merged['busiest_days']=result
```

#### Question 4: Create a new column containing these labels “Most orders,” “Average orders,” and “Fewest orders", and called “busiest_period_of_day.”

```python
# Which hour most orders take place
ords_prods_merged['order_hour_of_day'].value_counts(dropna = False)
```

#### Observations:  'Most orders' (hour 10, 11, 14, 15, 13, 12, 16, 9 - from 9am to 4pm), 'Average orders' (hour 17, 8, 18, 19, 20, 7, 21, 22 - from 7am to 8am and from 5pm to 10pm), or 'Fewest orders' (hour 23, 6, 0, 1, 5, 2, 4, 3 - from 11pm to 6am).

```python
result = []

for value in ords_prods_merged["order_hour_of_day"]:
  if value in[10, 11, 14, 15, 13, 12, 16, 9]:
    result.append("Most orders")
  elif value in[17, 8, 18, 19, 20, 7, 21, 22]:
    result.append("Average orders")
  else:
    result.append("Fewest orders")
```

```python
# Create a new column 'busiest_period_of_day'
ords_prods_merged['busiest_period_of_day']=result
```
